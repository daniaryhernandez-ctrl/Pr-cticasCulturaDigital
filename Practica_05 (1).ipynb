Recuerda:

## Ejercicio 2
Pide al usuario que ingrese un número. Si el número es **positivo** muestra 'Es positivo', si es **negativo** muestra 'Es negativo', si es **cero** muestra 'Es cero'.

Escribe tu código en la celda siguiente:

````python
# Escribe tu código aquí
v2 = int(input("Ingrese un número: "))
if v2 > 0:
    print("Es positivo")

elif v2 < 0:
    print("Es negativo")

else:
    print("Es cero")
````

## Ejercicio 4
Pide al usuario que ingrese su calificación (0 a 100). Si la calificación es **mayor o igual a 60**, imprime 'Aprobado'. De lo contrario, imprime 'Reprobado'.

Escribe tu código en la celda siguiente:

````python
# Escribe tu código aquí
v4 = int(input("Ingrese su calificación: "))
if v4 >= 60:
    print("Aprobado")

else:
    print("Reprobado")
````
